# Task 1: Reading and Inspection
### 1.1 Import the database
Import the India's Covid database and store it in variable covid. 

```python
covid=pd.read_csv("covid_india.csv",index_col=0)
covid
```

Inspect details of dataframe like shape, columns, etc.

# Task 2: Cleaning The Data
### 2.1 Inspecting null values

```python
covid[covid.isnull()==True].count()
```

#### As there are no null values, the dataframe is ready to use.

# Task 3: Data Analysis and Visualization
### 3.1 Create more useful data and add it to the dataframe

Create two columns recovery rate and death rate in the dataframe 

```python
covid['Death Rate']=covid['Deaths']/covid['Total Confirmed cases']*100
covid
```

```python
covid['Recovery Rate']=covid['Cured/Discharged/Migrated']/covid['Total Confirmed cases']*100
covid
```

### 3.2 Clean the code after modifying the dataframe

```python
covid.fillna(0)
```

### 3.3 Total confirmed cases and active cases

```python
Tc,Ac=covid['Total Confirmed cases'].sum(),covid['Active Cases'].sum()
print("Total Confirmed Cases are {} and Total Active Cases are {}.".format(Tc,Ac))
```

### 3.4 Total deaths and cured patients

```python
deaths,cured=covid['Deaths'].sum(),covid['Cured/Discharged/Migrated'].sum()
print("Total number of deaths are {} and Total cured patients are {}.".format(deaths,cured))
```

##  Visualizing India's Covid-19 Data

```python
cov=pd.DataFrame({"Total Number(in millions)":[Tc,deaths,cured,Ac]},index=['Total Cases','Deaths','Recovered','Active Cases'])
cov
```

```python
plt.figure(figsize=(18,7))
plt.title("India's Covid-19 Data",size=20)
sns.set_style("darkgrid")
sns.set_context(font_scale=1.25, rc={"grid.linewidth": 1.5})
sns.barplot(y=cov['Total Number(in millions)'],x=cov.index,data=cov,palette="tab10")
```

### 3.5 States with highest number of cases, deaths and recovered patients

```python
ss=covid.iloc[covid['Total Confirmed cases'].argmax()]["Name of State / UT"]
ds=covid.iloc[covid['Deaths'].argmax()]["Name of State / UT"]
rs=covid.iloc[covid['Cured/Discharged/Migrated'].argmax()]["Name of State / UT"]
print("State with most confirmed cases is {}.".format(ss))
print("State with most deaths is {}.".format(ds))
print("State with most deaths is {}.".format(rs))
```

## Visualizing Statewise Data
### Statewise number of cases

```python
plt.figure(figsize=(16,10))
plt.title("Statewise number of cases in India",size=25)
sns.set_context(font_scale=1.5, rc={"grid.linewidth": 1.5})
sns.barplot(x=covid['Total Confirmed cases'],y=covid['Name of State / UT'],data=covid,palette="autumn")
```

### Statewise number of deaths 

```python
plt.figure(figsize=(16,10))
plt.title("Statewise number of deaths in India",size=25)
sns.set_context(font_scale=1.5, rc={"grid.linewidth": 1.5})
sns.barplot(x=covid['Deaths'],y=covid['Name of State / UT'],data=covid,palette="inferno_r")
```

### Statewise number of recoveries 

```python
plt.figure(figsize=(16,10))
plt.title("Statewise number of recoveries in India",size=25)
sns.set_context(font_scale=1.5, rc={"grid.linewidth": 1.5})
sns.barplot(x=covid['Cured/Discharged/Migrated'],y=covid['Name of State / UT'],data=covid,palette="spring")
```
